```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("football.csv")
categorical_columns = df.select_dtypes(include=['object']).columns
df = pd.get_dummies(df, columns=categorical_columns, prefix=categorical_columns, prefix_sep='_')
df['goalkeeping_speed'].fillna(0, inplace=True)
columns_drop= ['sofifa_id','value_eur','wage_eur','age','club_team_id','league_level','club_jersey_number','nation_jersey_number','international_reputation','release_clause_eur']
df.drop(columns = columns_drop, axis=1, inplace=True)
```

```python
y = df['overall']
df.drop('overall', axis=1, inplace=True)
X=df
```

```python
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)
```

```python
rf_mse = RandomForestRegressor(n_estimators=10, criterion='squared_error', random_state=42)
rf_mse.fit(X_train, y_train)
y_pred_mse = rf_mse.predict(X_test)
mse_mse = mean_squared_error(y_test, y_pred_mse)
```

```python
rf_mae = RandomForestRegressor(n_estimators=30, max_depth=10, criterion='absolute_error',max_samples=10,random_state=2)
rf_mae.fit(X_train, y_train)
y_pred_mae = rf_mae.predict(X_test)
mae_mae = mean_absolute_error(y_test, y_pred_mae)
```

```python
print(f"Random Forest (MSE) - Test MSE: {mse_mse:.4f}")
print(f"Random Forest (MAE) - Test MAE: {mae_mae:.4f}")
```
