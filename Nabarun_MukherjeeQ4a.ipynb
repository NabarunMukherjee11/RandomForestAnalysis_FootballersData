```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("football.csv")
categorical_columns = df.select_dtypes(include=['object']).columns
df = pd.get_dummies(df, columns=categorical_columns, prefix=categorical_columns, prefix_sep='_')
df['goalkeeping_speed'].fillna(0, inplace=True)
columns_drop= ['sofifa_id','value_eur','wage_eur','age','club_team_id','league_level','club_jersey_number','nation_jersey_number','international_reputation','release_clause_eur']
df.drop(columns = columns_drop, axis=1, inplace=True)
```

```python
y = df['contribution_type']
df.drop('contribution_type', axis=1, inplace=True)
X=df
type(X)
```

```python
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)
```

```python
criteria = ['gini', 'entropy']
results = {}
for criterion in criteria:
    clf = RandomForestClassifier(n_estimators=100, criterion=criterion, random_state=42)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='weighted')
    results[criterion] = {'accuracy': accuracy, 'f1_score': f1}
```

```python
for criterion, scores in results.items():
    print(f'Quality Criterion: {criterion}')
    print(f'Test Accuracy: {scores["accuracy"]:.4f}')
    print(f'F1 Score: {scores["f1_score"]:.4f}')
    print()
```
